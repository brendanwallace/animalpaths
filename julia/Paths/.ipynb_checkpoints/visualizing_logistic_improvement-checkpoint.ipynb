```julia
import Paths
using Plots
```

```julia
sim = Paths.MakeSimulation(numWalkers=10, numLocations=10, X=100, Y=100,
		patchImprovement=0.05, patchRecovery=0.0005,
		scenario=Paths.RANDOM_FIXED, maxCost=2.0, logisticPatchImprovement=true)
```

```julia
UPF = 5

@gif for t in 1:100*UPF
    Paths.update!(sim)
    histogram([(sim.world.patches...)...])
end every UPF
```
