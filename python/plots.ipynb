```python
import numpy as np
import matplotlib.pyplot as plt

from paths import scenario
```

```python
T = 1000
UPF = 64

patch_sums = {}

for pi in [0.01, 0.02, 0.04]:
    patch_sums[pi] = {}
    for pr in [0.0001, 0.0002, 0.0004]:
        patch_sums[pi][pr] = []
        s = scenario.RandomFixedLocations(pr=pr, pi=pi)
        label=f'pr={pr}|pi={pi}'
        for t in range(T):
            print(f'\r{label}|{t}/{T}            ', end='')
            for u in range(UPF):
                s.update()
            patch_sums[pi][pr].append(s.world.patches.sum())
            
```

```python
for pi, pr_dict in patch_sums.items():
    for pr, sums in pr_dict.items():
        plt.plot(sums, label=f'pr={pr}|pi={pi}')
plt.legend()
plt.title("Total improved path over time")
plt.xlabel("Time (in frames)")
plt.ylabel("Total improved path")
```

```python
colors = {
    0.0025: "purple",
    0.005: "blue",
    0.01: "green",
    0.02: "yellow",
    0.01: "orange",
    0.04: "red",
}
for pi, pr_dict in patch_sums.items():
    for pr, sums in pr_dict.items():
        ratio = pi/pr
        plt.plot(sums, label=f'pr={pr}|pi={pi}|ratio={ratio}', linewidth=ratio/100, color="")
plt.legend()
plt.title("Total improved path over time")
plt.xlabel("Time (in frames)")
plt.ylabel("Total improved path")
```
