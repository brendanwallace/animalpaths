```julia
using Paths
using Plots
using LinearAlgebra
```

```julia
S=100

settings = Paths.Settings(
    searchStrategy=Paths.KANAI_SUZUKI,
    X=S, Y=S, maxCost=10.0, patchImprovement=0.15, patchRecovery=0.002,
    numLocations=10, recoveryLogic=Paths.LOGISTIC, improvementLogic=Paths.LOGISTIC)
sim = Paths.MakeSimulation(settings);

F = 20
UPF = 100
@gif for f in 1:F
    print("\r$(f)/$(F)")
    for u in 1:UPF
        Paths.update!(sim)
    end
    Paths.viz(sim)
end
```

```julia
graph = Paths.viz(sim)
savefig(graph, "quantify-angles-1.png")
```

```julia
snapshot = Paths.snapshot(sim, savepaths=true);
paths = snapshot.paths;
```

```julia
pp = plot([p[1] for p in paths[1]], [p[2] for p in paths[1]], lims=(1, S + 2), legend=false, aspect_ratio=1)
# scatter!([(l.position[1], l.position[2]) for l in sim.locations], label="", color="#ff7f0e")
for path in paths
    plot!([p[1] for p in path], [p[2] for p in path])
end
display(pp)
savefig(pp, "quantify-angles-2.png")
```

```julia
hs = reduce(vcat, [Paths.weightedHeadings(path) for path in paths])
histogram([h[1]/(2*π)*360 for h in hs], weights=[h[2] for h in hs], legend=nothing, xlabel="relative heading (degrees)")
```

```julia
# function anglesInPath(path::Paths.Path)::Array{Float64}
# for _p in [19]

path = paths[31]
steps = []
for i in 1:length(path)-1
    push!(steps, path[i+1] .- path[i])
end
angles = [atan(s[2], s[1])/(π) for s in steps]
lengths = [norm(s) for s in steps]
cumulative = [sum(lengths[1:i]) for i in 1:length(lengths)]
curve = plot([p[1] for p in path], [p[2] for p in path], xlim=(1, 100), ylim=(1, 102), aspect_ratio=1, color=3, legend=nothing)
display(curve)

savefig(curve, "quantify-angles-3.png")

weighteds = Paths.weightedHeadings(path)
hist = histogram(
    [w[1]/(2*π)*360 for w in weighteds],
    weights=[w[2] for w in weighteds], bins=50, legend=nothing,
    ylabel="distance traveled",
    xlabel="relative heading (degrees)"
)
savefig(hist, "quantify-angles-4.png")
display(hist)

# display(plot(angles, title="headings in radians", ylim=(-1, 1)))
# display(plot(cumulative, angles, title="headings", ylim=(-1, 1)))
# display(plot(cumulative, [h[1] for h in Paths.weightedHeadings(path)], title="headings in relative radians"))
```
